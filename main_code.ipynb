```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data Set

```python
data=pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
x=data.drop(["median_house_value"],axis=1)
y=data["median_house_value"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
train_data=x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,9))
```

```python
data = train_data.select_dtypes(include=['number'])
data.corr()
```

```python
plt.figure(figsize=(15,8))
data = train_data.select_dtypes(include=['number'])
sns.heatmap(data.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

```python
train_data['total_rooms']=np.log(train_data["total_rooms"]+1)
train_data['total_bedrooms']=np.log(train_data["total_bedrooms"]+1)
train_data["population"]=np.log(train_data["population"]+1)
train_data["households"]=np.log(train_data["households"]+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
if 'ocean_proximity' in train_data.columns:
    # Print the unique values and their counts
    print(train_data['ocean_proximity'].value_counts())
else:
    print("The 'ocean_proximity' column does not exist in the DataFrame.")
```

```python
train_data.ocean_proximity.value_counts()
```

```python

train_data_ocean_proximity_encoded = pd.get_dummies(train_data.ocean_proximity)
train_data_ocean_proximity_encoded = train_data_ocean_proximity_encoded.astype(int)
train_data = train_data.join(train_data_ocean_proximity_encoded).drop(columns=['ocean_proximity'],axis=1)
print(train_data)
```

```python
plt.figure(figsize=(15,8))
data = train_data.select_dtypes(include=['number'])
sns.heatmap(data.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude",y="longitude",data=train_data,hue="median_house_value",palette="coolwarm")
```
